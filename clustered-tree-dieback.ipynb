```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def prepare_df(path):
    
    df = pd.read_excel(path)
    df = df.rename(columns = {
        'LwConfEnv':'Confidence Env.',
        'HiConfEnv':'Confidence Env.'
    }).drop(columns = ['OBJECTID', 'DiffK'])
    
    df['Index'] =  df['ExpectedK'].round(0)
    df = df.set_index('Index')
    return df
```

```python
def create_plot(df):
    
    sns.lineplot(data=df, palette=['b', 'r', 'g'])
    plt.grid() 
    plt.xlim(df.index[0],df.index[-1])
    plt.xlabel("Dystans")
    plt.ylabel("L(d)")
    plt.text(-5, 135, "Skupienie",  color = 'b', fontstyle = 'italic')
    plt.text(90, -20 , "Rozproszenie", color = 'b', fontstyle = 'italic')
    
```

```python
df_2014 = create_plot(prepare_df('./data/Ripley K/martwe_drzewa_2014.xls'))
```
