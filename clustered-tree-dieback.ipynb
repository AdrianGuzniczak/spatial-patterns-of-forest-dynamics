```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def prepare_df(path):
    
    df = pd.read_excel(path)
    df = df.rename(columns = {
        'LwConfEnv':'Confidence Env.',
        'HiConfEnv':'Confidence Env.'
    }).drop(columns = ['OBJECTID', 'DiffK'])
    
    df['Index'] =  df['ExpectedK'].round(0)
    df = df.set_index('Index')
    return df
```
